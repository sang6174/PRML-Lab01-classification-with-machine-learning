```python
import numpy as np

X_train = np.load('X_train.npy')
y_train = np.load('y_train.npy')
X_test = np.load('X_test.npy')
y_test = np.load('y_test.npy')
```

```python
import pickle

with open('model_logistic_regression.pkl', 'rb') as file:
    model_lr = pickle.load(file)

print(model_lr)
```

```python
with open('model_svm_linear.pkl', 'rb') as file:
    model_svm_linear = pickle.load(file)

print(model_svm_linear)
```

```python
with open('model_svm_rbf.pkl', 'rb') as file:
    model_svm_rbf = pickle.load(file)

print(model_svm_rbf)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

y_pred_lr = model_lr.predict(X_test)
y_pred_svm_linear = model_svm_linear.predict(X_test)
y_pred_svm_rbf = model_svm_rbf.predict(X_test)
```

```python
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr, average='weighted')
recall_lr = recall_score(y_test, y_pred_lr, average='weighted')
f1_lr = f1_score(y_test, y_pred_lr, average='weighted')
```

```python
accuracy_svm_linear = accuracy_score(y_test, y_pred_svm_linear)
precision_svm_linear = precision_score(y_test, y_pred_svm_linear, average='weighted')
recall_svm_linear = recall_score(y_test, y_pred_svm_linear, average='weighted')
f1_svm_linear = f1_score(y_test, y_pred_svm_linear, average='weighted')
```

```python
accuracy_svm_rbf = accuracy_score(y_test, y_pred_svm_rbf)
precision_svm_rbf = precision_score(y_test, y_pred_svm_rbf, average='weighted')
recall_svm_rbf = recall_score(y_test, y_pred_svm_rbf, average='weighted')
f1_svm_abf = f1_score(y_test, y_pred_svm_rbf, average='weighted')
```

```python
import pandas as pd
```

```python
results = {
    'Model': ['Logistic Regression', 'SVM Linear', 'SVM RBF'],
    'Accuracy': [accuracy_lr, accuracy_svm_linear, accuracy_svm_rbf],
    'Precision': [precision_lr, precision_svm_linear, precision_svm_rbf],
    'Recall': [recall_lr, recall_svm_linear, recall_svm_rbf],
    'F1 Score': [f1_lr, f1_svm_linear, f1_svm_abf]
}

df_results = pd.DataFrame(results)
print(df_results)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_results.melt(id_vars='Model'), x='variable', y='value', hue='Model')
plt.title('Comparison of Model Performance Metrics')
plt.ylabel('Score')
plt.xlabel('Metrics')
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylim(0, 1)
plt.show()
```

```python
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("SVM Linear Accuracy:", accuracy_score(y_test, y_pred_svm_linear))
print("SVM RBF Accuracy:", accuracy_score(y_test, y_pred_svm_rbf))

print("Logistic Regression Report:\n", classification_report(y_test, y_pred_lr))
print("SVM Linear Report:\n", classification_report(y_test, y_pred_svm_linear))
print("SVM RBF Report:\n", classification_report(y_test, y_pred_svm_rbf))
```
